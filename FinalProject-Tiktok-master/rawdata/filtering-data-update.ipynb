### Cleaning and Filtering Scraped Data

```python
##Necessary modules 
import pandas as pd
import numpy as np 
import datetime
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from scipy import stats
```

```python
##Reading in our CSV files as dataframes, a comprehensive dataset will be formed as a dictionary below
hypehouse_df = pd.read_csv("thehypehouse.csv")
charlidamelio_df = pd.read_csv("charlidamelio.csv")
avani_df = pd.read_csv("avani.csv")
tonylopez_df = pd.read_csv("tonylopez.csv")
k0uvr_df = pd.read_csv("k0uvr.csv")
jackwright_df = pd.read_csv("jack.wright.csv")
jameswright_df = pd.read_csv("jameswrightt.csv")
ondreazlopez_df = pd.read_csv("ondreazlopez.csv")
addisonre_df = pd.read_csv("addisonre.csv")
alexwarren_df = pd.read_csv("alexwaarren.csv")
rylandstorms_df = pd.read_csv("rylandstormss.csv")
petroutv_df = pd.read_csv("petroutv.csv")
larrayeeee_df = pd.read_csv("larrayeeee.csv")
lilhuddy_df = pd.read_csv("lilhuddy.csv")
dixiedamelio_df = pd.read_csv("dixiedamelio.csv")
nickaustin_df = pd.read_csv("nickaustinn.csv")
```

```python
##Cleaning up Dataframe, dropping unnecessary columns 
hypehouse_members_data = [hypehouse_df, charlidamelio_df, avani_df, tonylopez_df, k0uvr_df, jackwright_df, jameswright_df, ondreazlopez_df, addisonre_df, alexwarren_df, rylandstorms_df, petroutv_df, larrayeeee_df, lilhuddy_df, dixiedamelio_df, nickaustin_df]
hypehouse_members_names = ["thehypehouse", "charlidamelio", "avani", "tonylopez", "k0uvr", "jackwright", "jameswright", "ondreazlopez", "addisonre", "alexwarren", "rylandstorms","petroutv", "larrayeeee", "lilhuddy", "dixiedamelio", "nickaustin"]

to_drop = ['id', 'authorMeta.id', 'authorMeta.nickName', 'authorMeta.following', 'authorMeta.fans', 'authorMeta.digg',
       'authorMeta.verified', 'authorMeta.private', 'authorMeta.signature',
       'authorMeta.avatar', 'musicMeta.musicId', 'musicMeta.musicOriginal', 'musicMeta.playUrl',
       'covers.default', 'covers.origin', 'covers.dynamic', 'imageUrl',
       'webVideoUrl', 'videoUrl', 'videoUrlNoWaterMark', 'videoMeta.width',
       'videoMeta.height', 'videoMeta.ratio', 'downloaded']

for x in range(len(hypehouse_members_data)):
    hypehouse_members_data[x] = hypehouse_members_data[x].drop(to_drop, axis = 1)
```

```python
##ref_date is date of the first post in the dataframe 
##user is a dataframe of the Tiktok user's information
##Dates column will be added for each dataframe, keeping createTimes columns for now in case needed in future, but as you can observe the format is very confusing
dates = [[5,4,2020], [5,5,2020], [5,5,2020], [5,5,2020], [5,4,2020], [5,5,2020], [5,4,2020], [5,4,2020], [5,5,2020], [5,3,2020], [5,4,2020], [5,5,2020], [5,3,2020], [5,4,2020], [5,3,2020], [5,4,2020]]

def createTime_to_date (user, ref_date):
    ref_date = datetime.date(ref_date[2], ref_date[0], ref_date[1])
    ref_time = user["createTime"][0]
    times = np.asarray(user['createTime'])
    days = np.round(np.abs(times - ref_time)/86400, 0)
    
    dates = []
    for x in days:
        conv_days = datetime.timedelta(x)
        date = ref_date - conv_days
        dates.append(date)
        
    dates - np.asarray(dates)
    
    return dates

for x in range(len(hypehouse_members_data)):
    post_dates = createTime_to_date(hypehouse_members_data[x], dates[x])
    hypehouse_members_data[x]["dates"] = post_dates 
```

```python
##Combining data into one accessible dictionary
dataset = {}
for member in range(len(hypehouse_members_data)):
    dataset.update({hypehouse_members_names[member]: hypehouse_members_data[member]})
```

```python
#creating a column in each dataframe representing Engagement Rate of Tiktoker's posts in percentage
for df in hypehouse_members_data:
    df["Engagement Rates"] = ((df["diggCount"] + df["commentCount"] + df["shareCount"]) / df["playCount"]) * 100
```

```python
#creating a new column in each dataframe 
for df in hypehouse_members_data:
    is_joined_hypehouse = df['dates'] >= datetime.date(2019, 12, 19)
    df["Hypehouse?"] = is_joined_hypehouse.astype(int)
```
